```python
import matplotlib.pyplot as plt

# Data
x = list(range(11))
delay = [4599.56, 3807.50, 4100.78, 3724.85, 3658.32, 3817.57, 3804.30, 3772.11, 3865.37, 3926.80, 4093.29]
lev = [274, 186, 137, 122, 108, 103, 101, 100, 99, 98, 98]

fig, ax1 = plt.subplots()

# Plotting delay
ax1.set_xlabel('Time of Operator Implementation')
ax1.set_ylabel('Delay', color='tab:blue')
ax1.plot(x, delay, color='tab:blue', label='Delay')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a twin Axes sharing the xaxis
ax2 = ax1.twinx()

# Plotting lev
ax2.set_ylabel('Lev', color='tab:red')
ax2.plot(x, lev, color='tab:red', label='Lev')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Adding title
plt.title('Delay and Lev over Time of Operator Implementation')

# Showing the plot
plt.show()

# save as pdf
fig.savefig('delay_lev.pdf', bbox_inches='tight')

# save as svg
fig.savefig('delay_lev.svg', bbox_inches='tight')
```

```python
import csv
import matplotlib.pyplot as plt

# 读取CSV文件
#csv_file = '/data/cchen/esyn2_new/E-syn2/data_analysis/Logadder_100.csv'
#csv_file = '/data/cchen/esyn2_new/E-syn2/data_analysis/adder_timing_100.csv'
# csv_file = '/data/cchen/esyn2_new/E-syn2/data_analysis/arb_timing_100.csv'
# csv_file = '/data/cchen/esyn2_new/E-syn2/data_analysis/adder.csv'
csv_file = 'max.csv'
#csv_file = '/data/cchen/esyn2_new/E-syn2/data_analysis/bar_100.csv'
iteration = []
delay = []
running_time = []

with open(csv_file, 'r') as file:
    reader = csv.reader(file)
    next(reader)  # 跳过标题行
    for row in reader:
        iteration.append(int(row[0]))
        delay.append(float(row[1]))
        running_time.append(float(row[3]))  # 添加Running time的值

# 创建图形和坐标轴对象
fig, ax1 = plt.subplots()

ax1.set_xlabel('Egraph Rewriting Iteration')
ax1.set_ylabel('Delay (ps)', color='tab:blue')
# 绘制Delay曲线（左坐标轴）
ax1.plot(iteration, delay, color='tab:blue')
# ax1.set_xlabel('Iteration')
# ax1.set_ylabel('Delay (ps)', color='blue')
ax1.tick_params('y', colors='tab:blue')

# 创建第二个坐标轴对象
ax2 = ax1.twinx()

# 绘制Running time曲线（右坐标轴）
ax2.set_ylabel('Running time (s)', color='tab:purple')
ax2.plot(iteration, running_time, color='tab:purple')
# ax2.set_ylabel('Running time (s)', color='red')
ax2.tick_params('y', colors='tab:purple')

# 添加标题和网格线
# plt.title('Delay and Running time vs. Iteration')
#plt.grid(True)

plt.title('Delay and Runtime vs. E-graph Rewriting Iteration')

# 显示图形
#plt.show()

# 保存图形为PNG文件
#plt.savefig('adder_timing_100.png')
#plt.savefig('Logadder_100.png')
#plt.savefig('max_100.svg')  # 修改这里保存为SVG格式
#plt.savefig('bar_100.png')
plt.savefig('max.pdf')  # 修改这里保存为PDF格式
```
